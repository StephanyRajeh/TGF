```python
def degree_distribution(G):
    vk = dict(G.degree())
    vk = list(vk.values()) # we get only the degree values
    maxk = np.max(vk)
    mink = np.min(min)
    kvalues= np.arange(0,maxk+1) # possible values of k
    Pk = np.zeros(maxk+1) # P(k)
    for k in vk:
        Pk[k] = Pk[k] + 1
    Pk = Pk/sum(Pk) # the sum of the elements of P(k) must to be equal to one
    return kvalues,Pk
```

```python
ks, Pk = degree_distribution(g)
```

```python
plt.figure()
plt.loglog(ks,Pk,'bo',basex=10,basey=10)
plt.xlabel("k", fontsize=20)
plt.ylabel("P(k)", fontsize=20)
plt.title("Degree distribution", fontsize=20)
plt.grid(True)
#plt.savefig('degree_dist.eps') #save the figure into a file
plt.show(True)
```

```python
def momment_of_degree_distribution(G,m):
    k,Pk = degree_distribution(G)
    M = sum((k**m)*Pk)
    return M
```

```python
k1 = momment_of_degree_distribution(g,1)
print("Mean degree = ", round(np.mean(vk),3))
print("First moment of the degree distribution = ", round(k1,3))
```

```python
k2 = momment_of_degree_distribution(g,2)
print("Second moment of the degree distribution = ", round(k2,3))
```

The variance $V(k) = \langle k^2 \rangle - k^2$

```python
variance = momment_of_degree_distribution(g,2) - momment_of_degree_distribution(g,1)**2
print("Variance of the degree = ", round(variance,3))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import auc
from sklearn.metrics import RocCurveDisplay
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold


clf = RandomForestClassifier(random_state=12)
cv = 10

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

fig, ax = plt.subplots(figsize=(10, 10))

for i, (train, test) in enumerate(StratifiedKFold(n_splits=cv).split(X_ndarray, y_ndarray)):
    clf.fit(X_ndarray[train], y_ndarray[train])
    viz = RocCurveDisplay.from_estimator(clf, X_ndarray[test], y_ndarray[test], ax=ax, name='ROC fold {}'.format(i))
    interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)

ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r', alpha=.8)
mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)

ax.plot(mean_fpr, mean_tpr, color='b', label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc), lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)

ax.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2, label=r'$\pm$ 1 std. dev.')

ax.set(xlim=[-0.05, 1.05], ylim=[-0.05, 1.05], title="Receiver operating characteristic example")
ax.legend(loc="lower right")

plt.show()
```

### Interpretation using permutation feature importance

Permutation feature importance measures the importance of a feature by looking at how much the error increases after we change the feature values. The feature is important if the change of its value increases the model’s error.

To implement this method, we can use a function called permutation_importance from the scikit-learn library to calculate the feature importance. From that result, we will create a boxplot to visualize the feature importance.

Source: https://towardsdatascience.com/how-to-remove-multicollinearity-using-python-4da8d9d8abb2

```python
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 12)

print(X.shape, X_train.shape, X_test.shape)
```

```python
model = RandomForestClassifier(random_state=12)
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
result = permutation_importance(model, X_train, Y_train, n_repeats=10, random_state=12)

perm_imp_idx = result.importances_mean.argsort()
plt.boxplot(result.importances[perm_imp_idx].T, vert=False,
            labels=X.columns[perm_imp_idx])
plt.title('Feature Importance in UCI-H1000-O1 with RandomForestClassifier', fontweight="bold")
plt.show()
```

### Interpretation using the returned estimators

```python
# https://stackoverflow.com/questions/55466081/how-to-calculate-feature-importance-in-each-models-of-cross-validation-in-sklear
from sklearn.model_selection import cross_validate
import random

clf=RandomForestClassifier(random_state=12)
output = cross_validate(clf, X, Y, cv=10, scoring = 'roc_auc', return_estimator =True)

df = pd.DataFrame()
for idx,estimator in enumerate(output['estimator']):
    print("Features sorted by their score for estimator {}:".format(idx))
    feature_importances = pd.DataFrame(estimator.feature_importances_,
                                       index=X.columns,
                                        columns=['importance']).sort_values('importance', ascending=False)
    df = pd.concat([df, feature_importances], axis=1)
```
